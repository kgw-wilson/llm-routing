## Plotting Experiment Results

```python
import matplotlib.pyplot as plt

k_vals = [5, 10, 40]
k_labels = [str(k) for k in k_vals]
x = range(len(k_vals))  # evenly spaced positions

experiments = {
    "exp1 (Prompt Embedding Space)": [0.7706849315068494, 0.806027397260274, 0.8476712328767123],
    "exp2 (Response Embedding Space)": [0.7435616438356164, 0.8135159817351598, 0.8523287671232876],
    "exp3 (Concatenation)": [0.7732420091324201, 0.8084018264840183, 0.8491324200913242],
    "exp4 (Difference)": [0.7734246575342466, 0.8076712328767123, 0.8484018264840183],
    "exp5 (Weighted difference, =0.5, =0.5)": [0.7734246575342466, 0.8076712328767123, 0.8484018264840183],
    "exp6 (Prompt angle space)": [0.7484931506849315, 0.7977168949771689, 0.8417351598173516],
    "exp7 (Response embedding smoothing from prompt neighbors)": [0.70675799086758, 0.8043835616438356, 0.8517808219178082],
    "exp9 (Task description embeddings)": [0.7665753424657534, 0.8043835616438356, 0.8461187214611872],
}

plt.figure(figsize=(9, 5))

for label, accuracies in experiments.items():
    plt.plot(x, accuracies, marker="o", label=label)

plt.xticks(x, k_labels)
plt.xlabel("k")
plt.ylabel("Accuracy")
plt.ylim(0.70, 0.86) 
plt.title("Accuracy vs k Across Experiments")
plt.legend(fontsize=9)
plt.grid(True, alpha=0.3)
plt.tight_layout()
```

```python
baseline = experiments["exp1 (Prompt Embedding Space)"]

plt.figure(figsize=(9, 5))

for label, accuracies in experiments.items():
    delta = [a - b for a, b in zip(accuracies, baseline)]
    plt.plot(x, delta, marker="o", label=label)

plt.axhline(0, linestyle="--", linewidth=1)
plt.xticks(x, k_labels)
plt.xlabel("k")
plt.ylabel(" Accuracy vs Prompt Embedding")
plt.title("Relative Accuracy Improvements")
plt.legend(fontsize=9)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
